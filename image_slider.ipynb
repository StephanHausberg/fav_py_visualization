# Imageslider

Adopting Image Slider for regular images to have a stellar presentation of your graphs.

### Create graphs using pandas

Random data is sent to two images.

```python
import pandas as pd


index = ['Dairy', 'Meat', 'Seafood', 'Bakery', 'Soft Drinks', 'Deli', 'Canned Goods']
df1 = pd.DataFrame(data = {'revenue 2023' : [40, 20, 10, 20, 10, 20, 30],
                           'revenue 2024' : [20, 15, 25, 23, 14, 25, 35]},
        index = index)
ax = df1.plot.bar(rot=0, legend = False)
fig = ax.get_figure()
fig.savefig("plot01.png")

df2 = pd.DataFrame(data = {'revenue 2023' : [40, 20, 10, 20, 10, 20, 30],
                           'revenue 2024' : [0, 0, 0, 0, 0, 0, 0]},
        index = index)
ax = df2.plot.bar(rot=0, legend = False)
fig = ax.get_figure()
fig.savefig("plot02.png")
```

### Create Image Slider

```python
from d3blocks import D3Blocks

d3 = D3Blocks()

img_before = "/Users/stephanhausberg/Documents/python/basic_work/fav_py_visualization/plot01.png"
img_after = "/Users/stephanhausberg/Documents/python/basic_work/fav_py_visualization/plot02.png"
d3.imageslider(img_before,
                      img_after,
                      showfig=True,
                      scale=True,
                      figsize=[400, 400],
                      notebook = True,
                      filepath = "/Users/stephanhausberg/Documents/python/basic_work/fav_py_visualization/image_slider_out.html")
```
