```python
import bar_chart_race as bcr
import pandas as pd

# start out creating a DataFrame in Pandas

sampl1 = [10000, 11000, 12000, 12000, 13000, 14000, 15000, 15000, 16000, 17000]
sampl2 = [0, 9000, 9000, 10000, 10000, 10000, 9000, 10000, 10000, 10000]
sampl3 = [0, 0, 9000, 9500, 10000, 10500, 12000, 12500, 13000, 13500]
sampl4 = [0, 0, 0, 8500, 8500, 8500, 8500, 8500, 8500, 8500]
sampl5 = [0, 0, 0, 0, 8500, 10000, 11500, 13000, 14500, 16000]
sampl6 = [0, 0, 0, 0, 0, 7000, 7000, 7000, 7000, 7000]

df = pd.DataFrame(data = {'01_process_start' : sampl1,
                         '02_process_step1': sampl2,
                         '03_process_step2': sampl3,
                         '04_process_step3': sampl4,
                         '05_process_step4': sampl5,
                         '06_process_step5': sampl6})

df = df.reset_index()

df["date"] = ["2022-12-01", "2022-12-02", "2022-12-03",
              "2022-12-04", "2022-12-05", "2022-12-06",
              "2022-12-07", "2022-12-08", "2022-12-09",
              "2022-12-10"]

df = df.set_index('date')
```

```python
df
```

### Create a bar chart race file

By applying the bar chart race package, we take the aforementioned DataFrame and define an outut file called 'process.mp4'. There are a lot of parameters for fine tuning and experimentation here. Find details for customization on <A HREF="https://pypi.org/project/bar-chart-race/">Bar Chart Race Webpage</A>.

```python
bcr.bar_chart_race(
    df=df,
    filename='process.mp4',
    orientation='h',
    sort='desc',
    n_bars=5,
    fixed_order=df.columns.to_list(),
    fixed_max=True,
    steps_per_period=20,
    interpolate_period=False,
    label_bars=True,
    bar_size=.95,
    figsize=(5, 3),
    dpi=144,
    cmap='dark12',
    title='Process evolution in time',
    title_size='',
    bar_label_size=7,
    tick_label_size=7,
    shared_fontdict={'family' : 'Arial', 'color' : '.1'},
    scale='linear',
    writer=None,
    fig=None,
    bar_kwargs={'alpha': .7},
    filter_column_colors=False) 
```
