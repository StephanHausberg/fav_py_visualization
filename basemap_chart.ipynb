## Create a basic Basemap

First, we import the corresponding packages that we need in this case. That will be Basemap and matplotlib for setting the sizes of the picture correctly.

```python
# libraries
from mpl_toolkits.basemap import Basemap

import matplotlib.pyplot as plt
```

With this at hand we are able to create a specific map. In this case we specify which rectangle of the world we want to take a look at. Later on, there are parameters specifying the colors of the map.

Particularly hard to find are the points from different cities in the world - google maps helps here in a way. Finally, we cycle through these coordinates to draw the Great Arcs onto a map.

```python
m=Basemap(llcrnrlon=-20,
          llcrnrlat=0,
          urcrnrlon=55,
          urcrnrlat=65,
          projection='merc')
plt.rcParams["figure.figsize"]=15,12

m.drawmapboundary(fill_color='lightblue',
                  linewidth=0)
m.fillcontinents(color='grey',
                 alpha=0.7,
                 lake_color='grey')
m.drawcoastlines(linewidth=0.1,
                 color="white")

# Add a connection between new york and London
startlat = 51.222983; startlon = 6.781065

lat_lon_list = [[51.53, 0.08], # London
               [52.24,21.019], # Warschawa
               [59.30, 17.97], # Stockholm
               [45.450, 9.170162], # Milano
               [34.85, 33.253682]] # Cyprus
for list_item in lat_lon_list:
    m.drawgreatcircle(startlon,
                      startlat,
                      list_item[1],
                      list_item[0],
                      linewidth=2,
                      color='darkblue');
```
